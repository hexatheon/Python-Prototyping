```python
def extract_from_dict(dictionary, key):
    if dictionary is None:
        return pd.DataFrame()
    else:
        return dictionary[key]
```

```python
timeDiffPrevWeek = pd.DataFrame()
tradeSumPrevWeek = pd.DataFrame()
bookSumPrevWeek = pd.DataFrame()
rollingPrevWeek = pd.DataFrame()

now = datetime.datetime.today()
now = '2016-02-11' # Warning: this line purely for testing
now = pd.Timestamp(now, tz=tz)
date = ExchangeInfo.roundTimestampToTradeDate(exchange, now)
bizrange = pd.bdate_range(end = date, periods = 6, tz = tz)

reader = CmePacketStatisticsReportDataReader(config_file=config_file)
for i in range(0,5):  
	tmp_stats = reader.read(trade_date = bizrange[i])
	timeDiffPrevWeek = pd.concat((timeDiffPrevWeek, extract_from_dict(tmp_stats, "timeDiffSummary")))
	tradeSumPrevWeek = pd.concat((tradeSumPrevWeek, extract_from_dict(tmp_stats, "tradeSumSummary")))
	bookSumPrevWeek = pd.concat((bookSumPrevWeek, extract_from_dict(tmp_stats, "bookSumSummary")))
	rollingPrevWeek = pd.concat((rollingPrevWeek, extract_from_dict(tmp_stats, "rolling")))
    
timeDiffPrevWeek = timeDiffPrevWeek.groupby(timeDiffPrevWeek.index).mean()
tradeSumPrevWeek = tradeSumPrevWeek.groupby(tradeSumPrevWeek.index).mean()
bookSumPrevWeek = bookSumPrevWeek.groupby(bookSumPrevWeek.index).mean()
#rollingPrevWeek = rollingPrevWeek.groupby(rollingPrevWeek.index).mean()
```

## Trade Summaries & Details Sum/Packet

```python
tradeCombined = pd.merge(left = packet_stats["tradeSumSummary"], 
                        right = tradeSumPrevWeek, 
                        how = 'outer', left_index = True, right_index = True)
tradeCombined = tradeCombined.rename(columns = {'frequency_x':'current', 'frequency_y':'prevWeek'})
tradeCombined = pd.DataFrame({'tradeSize': tradeCombined.index,
                              'current':tradeCombined.current,
                              'prevWeek': tradeCombined.prevWeek}).reindex_axis(['tradeSize', 'current', 'prevWeek'],
                                                                                 axis = 1)
tradeCombined.index.name = None
#print tradeCombined
Latex(fancify_table(tradeCombined.to_latex(index = False)))
```

```python
fig, ax = plt.subplots()
width = 0.5
bar1 = ax.bar(packet_stats["tradeSumSummary"].index, packet_stats["tradeSumSummary"].frequency, width = width, color='b')
bar2 = ax.bar(tradeSumPrevWeek.index + 0.5, tradeSumPrevWeek.frequency, width = width, color='r')
ax.set_xlabel('tradeSize')
ax.set_ylabel('Frequency')
ax.set_title('TradeSummaries + TradeDetails per Packet (current vs PrevOneWeekAverage)', fontsize = 14)
ax.legend((bar1[0], bar2[0]), ('Current', 'PrevOneWeekAverage'))
plt.gcf().set_size_inches(20, 15)
```
